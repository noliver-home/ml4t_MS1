```python
#QuandlSimplify
#zipline_test
#This notebook is deliverable #2 for ML4T Milestone 1
#Milestone 1 takes me through installing miniconda, installing zipline, loading data from Quandl, and
#running a simple trading tutorial (Dual Moving Average).
#See https://zipline.ml4trading.io/beginner-tutorial
#and https://liveproject.manning.com/module/199_2_1/algorithmic-trading-with-machine-learning/1--setting-up-your-environment-and-sourcing-the-data/1-1-workflow%3a-setting-up-your-environment-and-sourcing-the-data?
#Assume that data from Quandl has been downloaded "out of band", i.e., after getting zipline installed, running
#zipline ingest -b Quandl
#This data is stored under C:\Users\<me>\.zipline 

#This deliverable alows me to either download a file us_stocks.csv from NASDAQ QUANDL database, or to download it via APIs.
#I have chosen to download it directly.
#The file is "WIKI_PRICES_212b326a081eacca455e13140d7bb9db.csv"  on my system
#After that, there are several simplifications to make:
#Change the date column to datetime format
#select only data from 2000 onward
#set "ticker" and "data" as indices
#Get rid of the original open, low, high, close, volume colums
#Rename the adj_ version of those columns (e.g., adj_open --> open)
#Save the frame as an HDF5 file


from zipline.api import order_target, record, symbol
import pandas as pd
import matplotlib.pyplot as plt

#The file that I downloaded from NASDAQ was specifically named "WIKI_PRICES_212b326a081eacca455e13140d7bb9db.csv"
wikidata = pd.read_csv("WIKI_PRICES_212b326a081eacca455e13140d7bb9db.csv")

#DEBUG - look at the data
#wikidata.head()

#Change date to datetime format
wikidata['date'] = pd.to_datetime(wikidata['date'])

#DEBUG - look at the data
#wikidata.head()


#Select data after 2000
filtered_wikidata = wikidata.loc[wikidata['date'] >= '2000-01-01']

#DEBUG - look at the data
#filtered_wikidata.head()

#Set indices
filtered_wikidata.set_index('ticker', 'date')

#DEBUG - look at the data
#filtered_wikidata.head()

#Remove the columns that are not required by the assignment
filtered_wikidata = filtered_wikidata.drop({'open', 'close', 'high', 'low', 'volume', 'ex-dividend', 'split_ratio'}, 1)

#DEBUG - look at the data
#filtered_wikidata.head()

#Rename the columns
filtered_wikidata.rename(columns = {'adj_open':'open', 'adj_high':'high', 'adj_low':'low', 'adj_close':'close', 'adj_volume':'volume'}, inplace = True)

#DEBUG - Look at the data
#filtered_wikidata.head()

#Save the frame in HD5
filtered_wikidata.to_hdf('us_stocks1.h5', key='us_stocks', mode='w')


```
